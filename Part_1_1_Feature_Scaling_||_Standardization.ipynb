#Example 1

```python
#Loading the libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
age = np.random.randint(20, 60, 100)  # Generate random ages between 20 and 60
salary = np.random.randint(30000, 100000, 100)  # Generate random salaries between 30000 and 100000

# Create DataFrame
data = {
    'Age': age,
    'Salary': salary
}
df = pd.DataFrame(data)
df.head()
```

```python
# Select columns to be standardized
cols_to_standardize = ['Age', 'Salary']

# Initialize StandardScaler
scaler = StandardScaler()

# Fit and transform the data
df_scaled = scaler.fit_transform(df[cols_to_standardize])

# Convert scaled data back to DataFrame
df_scaled = pd.DataFrame(df_scaled, columns=cols_to_standardize)
np.round(df_scaled.head(), 1)
```

```python
np.round(df_scaled.describe(), 1)
```

```python
# Plotting before and after scaling
plt.figure(figsize=(10, 5))

# Before scaling
plt.subplot(1, 2, 1)
plt.scatter(df['Age'], df['Salary'], color='blue', alpha=1)
plt.title('Before Scaling')
plt.xlabel('Age')
plt.ylabel('Salary')

# After scaling
plt.subplot(1, 2, 2)
plt.scatter(df_scaled['Age'], df_scaled['Salary'], color='red', alpha=1)
plt.title('After Scaling')
plt.xlabel('Age (Standardized)')
plt.ylabel('Salary (Standardized)')

plt.tight_layout()
plt.show()
```
